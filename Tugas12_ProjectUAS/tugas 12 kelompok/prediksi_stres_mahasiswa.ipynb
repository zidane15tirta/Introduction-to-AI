# Analisis dan Prediksi Tingkat Stres Mahasiswa
**Mata Kuliah:** Introduction to Artificial Intelligence (INF212)
**Dataset:** 94 Responden Mahasiswa Universitas Pembangunan Jaya

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report
import warnings

warnings.filterwarnings('ignore')
print("Library berhasil di-import!")
```

```python
nama_file = 'Responden Tingkat stress mhs - Form Responses 1 (1).csv'
df = pd.read_csv(nama_file)

print(f"Total Data Responden: {df.shape} baris, {df.shape} kolom\n")
df.head()
```

```python
X = df.iloc[:, 4:10] # Mengambil kriteria: Semester, Jam Tidur, Jumlah Tugas, Kelelahan, Kecemasan, Organisasi
y = df.iloc[:, 10]  # Mengambil target: Tingkat Stres

print("Kriteria Fitur (X):")
print(X.columns.tolist())
print(f"\nTarget Prediksi (y): {y.name}")
```

```python
X_encoded = pd.DataFrame()
le_features = {} 

for col in X.columns:
    le = LabelEncoder()
    X_encoded[col] = le.fit_transform(X[col].astype(str))
    le_features[col] = le

le_target = LabelEncoder()
y_encoded = le_target.fit_transform(y.astype(str))

print("Proses encoding dari teks ke angka berhasil dilakukan!")
print(f"Kategori kelas pada target: {le_target.classes_}")
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y_encoded, test_size=0.2, random_state=42)

print(f"Jumlah Data Latih (Training): {X_train.shape} baris")
print(f"Jumlah Data Uji (Testing)  : {X_test.shape} baris")
```

```python
model_nb = GaussianNB()
model_nb.fit(X_train, y_train)

print("Model Naive Bayes berhasil dibuat dan dilatih menggunakan 80% data latih!")
```

```python
from sklearn.tree import DecisionTreeClassifier

# 1. Inisialisasi dan Latih Model Kedua (Decision Tree)
model_dt = DecisionTreeClassifier(random_state=42)
model_dt.fit(X_train, y_train)

# 2. Prediksi Menggunakan Kedua Model
y_pred_nb = model_nb.predict(X_test)
y_pred_dt = model_dt.predict(X_test)

# 3. Hitung Akurasi
akurasi_nb = accuracy_score(y_test, y_pred_nb) * 100
akurasi_dt = accuracy_score(y_test, y_pred_dt) * 100

print("="*65)
print("          PERBANDINGAN AKURASI MODEL MACHINE LEARNING          ")
print("="*65)
print(f"1. Akurasi Model Naive Bayes   : {akurasi_nb:.2f}%")
print(f"2. Akurasi Model Decision Tree : {akurasi_dt:.2f}%")
print("-" * 65)

print("\n[DETAIL PERFORMANCE: NAIVE BAYES]")
print(classification_report(y_test, y_pred_nb, target_names=le_target.classes_, zero_division=0))

print("\n[DETAIL PERFORMANCE: DECISION TREE]")
print(classification_report(y_test, y_pred_dt, target_names=le_target.classes_, zero_division=0))
print("="*65)
```

```python
# Format input 6 kriteria: [[Semester, Tidur, Tugas, Lelah, Cemas, Organisasi]]
data_baru = [[4, 6, 8, 3, 4, 1]]  # <-- isi nilai yang ingin diprediksi

# Eksekusi Prediksi dari Kedua Model
angka_nb = model_nb.predict(data_baru)
angka_dt = model_dt.predict(data_baru)

teks_nb = le_target.classes_[angka_nb]
teks_dt = le_target.classes_[angka_dt]

print("="*65)
print("         PERBANDINGAN HASIL PREDIKSI DATA MAHASISWA BARU        ")
print("="*65)
print(f"Data Kriteria Input                : {data_baru[0]}")
print(f"Hasil Prediksi AI (Naive Bayes)    : Tingkat Stres -> {teks_nb[0]}")
print(f"Hasil Prediksi AI (Decision Tree)  : Tingkat Stres -> {teks_dt[0]}")
print("="*65)
```
